```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('database.csv').drop(columns='Unnamed: 0')
```

```python
df = df.dropna()
```

```python
df
```

```python
df.columns
```

```python
m_features = df[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo','duration_ms']]
```

```python
m_features
```

```python
face_carac = df[['age', 'gender', 'angry',
       'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']]
```

```python
face_carac['gender'] = face_carac['gender'].replace('Man',0).replace('Woman',1)
```

```python
face_carac
```

```python
scaler = StandardScaler()

scaled = scaler.fit_transform(face_carac)

scaled_face_carac = pd.DataFrame(scaled, index=face_carac.index, columns=face_carac.columns)

print(scaler.mean_)
```

```python
scaled_face_carac
```

```python
df_train = scaled_face_carac.join(m_features)
```

```python
df_train
```

```python
from sklearn.neighbors import NearestNeighbors

X = df_train

n_neighbors = 5

knn = NearestNeighbors(n_neighbors=n_neighbors+1, metric= "cosine")

knn.fit(X)

# Select a song from a user
item_selected = df_train.iloc[[0]]
#item_selected = scaled_musics[ scaled_musics.id== "1ZJ4MnRUj3gVg9AEFPZSi7" ]
#display(scaled_musics.id)
display(df_train.iloc[[0]])

item_selected = item_selected[X.columns] 

# Determine the neighbors
d, neighbors = knn.kneighbors(item_selected.values.reshape(1, -1))
neighbors = neighbors[0][1::] 
d = d[0][1::] 

#neighbors_names = scaled_musics.iloc[ neighbors ].id
```

```python
neighbors_df = pd.DataFrame(columns=df_train.columns)
for n in neighbors:
  neighbors_df = neighbors_df.append(df_train.iloc[[n]], ignore_index=True)
  print(df_train.iloc[[n]])
  print('---------------------------------------------------------')
```

```python
neighbors_df
```

```python
neighbors_df = neighbors_df.drop(columns=face_carac.columns)
```
